```python
# Calculate PI using Monte Carlo (Thanks, Lin!)
import random, math, PIL, IPython.display
from PIL import ImageDraw

def approximate_pi(trials):
    "Approximate Pi using Monte Carlo analysis. Takes trials: number of random samples"

    xmax = 300 # Image X pixels
    ymax = 300 # Image Y pixels

    inside = 0
    img = PIL.Image.new('RGB', (xmax, ymax), "black")
    draw = ImageDraw.Draw(img)
    for i in range(trials):
        x = random.random()
        y = random.random()
        # Euclidean distance, sqrt(x*x + y*y).
        if math.hypot(x, y) < 1:
            inside += 1
            draw.point((int(x*xmax), int(y*ymax)), fill="green")
        else:
            draw.point((int(x*xmax), int(y*ymax)), fill="red")
    img.save('monte-carlo-pi.png')

    # We've drawn 1/4 of the circle! Multiply by 4!
    pi = 4.0 * inside / trials
    return ('trials = {}, π ~= {}').format(trials, pi)
```

```python
approximate_pi(100000)
```

```python
IPython.display.Image(filename='monte-carlo-pi.png')
```

```python
approximate_pi(10**6)
```
